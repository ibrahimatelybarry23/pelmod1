## 1. Struttura di una Classe

Una classe in C++ è divisa in due sezioni:

- **`public`**: metodi e dati accessibili da qualsiasi parte del programma
- **`private`**: metodi e dati accessibili **solo** dall'interno della classe stessa

La classe `Foo` ha:
- Un campo privato `int* x` — un **puntatore** a intero allocato sullo heap
- Due metodi pubblici: `print()` e `setX()`
- Un costruttore di default e un distruttore

> 📌 All'interno della classe facciamo solo la **dichiarazione** dei metodi (il prototipo).
> L'**implementazione** (il corpo) viene scritta fuori, usando la sintassi `NomeClasse::metodo()`.

````python
#include <iostream>
#include <vector>

class Foo {
public:
    Foo();            // costruttore di default
    ~Foo();           // distruttore
    void print();     // stampa lo stato dell'oggetto
    void setX(int a); // imposta lo stato dell'oggetto
private:
    int* x;  // ← accessibile SOLO dall'interno della classe
};
````

## 2. Costruttore di Default

Il **costruttore di default** viene chiamato ogni volta che creiamo un oggetto senza argomenti:

```cpp
Foo myObj;        // ← costruttore di default chiamato qui
Foo arr[10];      // ← chiamato 10 volte!
```

### Cosa fa step by step:
1. `x = new int` → alloca **4 byte** nello **heap** e salva l'indirizzo in `x`
2. `*x = 0` → scrive il valore `0` nella locazione di memoria puntata

### Stack vs Heap — perché usiamo `new`?

| | Stack | Heap |
|---|---|---|
| Gestione | Automatica | **Manuale** (`new` / `delete`) |
| Durata | Fino alla fine del blocco `{}` | Fino a `delete` |
| Dimensione | Limitata (~1–8 MB) | Molto grande |
| Quando usarlo | Dati piccoli, durata nota | Dati grandi, durata variabile |

````python
Foo::Foo() {
    x = new int;  // alloca un int nello HEAP
    *x = 0;       // inizializza a 0
    std::cout << "[Costruttore] oggetto creato @ " << x << std::endl;
}
````

## 3. Distruttore

Il **distruttore** viene chiamato automaticamente quando l'oggetto **esce dallo scope**:

```cpp
{
    Foo f;   // costruttore chiamato
    // ...
}            // ← distruttore chiamato automaticamente qui
```

### Perché il distruttore è fondamentale?

Il costruttore ha allocato memoria con `new`. Se non la liberiamo con `delete` nel distruttore,
quella memoria rimane occupata per tutta la durata del programma → **memory leak** 🕳️

> 🔑 **Regola generale**: ogni `new` deve avere il suo corrispondente `delete`.

### ⚠️ Cosa succede dopo `delete`?
Il puntatore `x` diventa un **dangling pointer** — punta a memoria liberata.
Usarlo dopo il `delete` è **undefined behavior** (crash o dati corrotti).

````python
Foo::~Foo() {
    std::cout << "[Distruttore] *x = " << *x << ", libero memoria @ " << x << std::endl;
    delete x;  // OBBLIGATORIO: libera la memoria allocata con new
    // dopo questo punto x è un dangling pointer — non usarlo!
}
````

## 4. Implementazione dei Metodi

- `print()` — stampa il valore puntato da `x` (usando `*x` per dereferenziare)
- `setX(int a)` — modifica il valore puntato da `x`

> 💡 Con `x` stamperemmo l'**indirizzo** di memoria (es. `0x7ffee4b`). Con `*x` stampiamo il **valore** (es. `42`).

````python
void Foo::print() {
    std::cout << "[print] valore = " << *x << " (indirizzo heap: " << x << ")" << std::endl;
}

void Foo::setX(int a) {
    *x = a;  // dereferenzia e scrive il nuovo valore
}
````

## 5. Accesso ai Membri Privati — Errore di Compilazione

Il campo `x` è **privato**: solo i metodi della classe possono accedervi.
Tentare di accedere a `x` dall'esterno causa un **errore di compilazione**:

```cpp
Foo myv2;
myv2.x = 3;  // ❌ ERRORE: 'x' is a private member of 'Foo'
```

Questo è il principio dell'**incapsulamento**: l'utente della classe non deve conoscere
i dettagli interni — interagisce solo tramite l'interfaccia pubblica (`setX`, `print`).

```
┌─────────────────── Foo ───────────────────┐
│  PUBLIC  ← accesso dall'esterno ✅        │
│    print(), setX()                        │
│  ─────────────────────────────────────── │
│  PRIVATE ← accesso solo dall'interno ✅   │
│    int* x                                 │
└───────────────────────────────────────────┘
```

````python
// Uso CORRETTO della classe:
{
    Foo myv1, myv2;

    myv1.setX(12);   // ✅ setX è pubblico
    myv2.setX(23);   // ✅

    myv1.print();
    myv2.print();

    // myv2.x = 3;   // ❌ ERRORE: x è privato → decommentare per vedere l'errore
}
````

## 6. Oggetti Globali e Ordine di Inizializzazione

Considera questo codice:

```cpp
int prova() {
    globale.setX(10);  // usa 'globale'
}
Foo globale;  // dichiarata DOPO prova()
```

In C++, le **variabili globali** vengono inizializzate prima del `main()`, 
ma l'**ordine** tra file diversi è indefinito.
Se `prova()` venisse chiamata prima che `globale` fosse inizializzata → **undefined behavior**.

> 🔑 **Best practice**: evita variabili globali con costruttori non banali.
> Preferisci lo **Singleton pattern** o il passaggio per parametro.

## 7. Oggetti Dinamici — `new` e `delete`

Possiamo allocare oggetti nello **heap** con `new`.
In questo caso otteniamo un **puntatore** all'oggetto, e usiamo `->` invece di `.`:

```cpp
Foo obj;    // stack → obj.setX(5)   // operatore punto
Foo* pf;    // heap  → pf->setX(5)   // operatore freccia
```

| Sintassi | Quando si usa |
|---|---|
| `obj.metodo()` | L'oggetto è **diretto** (stack o riferimento) |
| `pf->metodo()` | L'oggetto è **puntato** (puntatore) |

### ⚠️ Con oggetti dinamici il distruttore NON è automatico!
Il distruttore viene chiamato solo quando scrivi **esplicitamente** `delete pf`.

````python
{
    std::cout << "--- Oggetto dinamico ---" << std::endl;

    Foo* pf = new Foo;  // costruttore chiamato, oggetto NELLO HEAP
    pf->setX(100);      // usa -> perché pf è un PUNTATORE
    pf->print();

    delete pf;          // distruttore chiamato ESPLICITAMENTE
    // pf ora è un dangling pointer — non usarlo più!
    pf = nullptr;       // buona pratica: resetta il puntatore
}
````

## 8. Array di Oggetti

Quando dichiariamo un array di oggetti, il costruttore viene chiamato **N volte** (una per elemento).
Alla fine dello scope, il distruttore viene chiamato **N volte** in **ordine inverso** (LIFO).

```
Foo myv[3]:
  → costruttore chiamato per myv[0]
  → costruttore chiamato per myv[1]
  → costruttore chiamato per myv[2]
  ... fine scope ...
  → distruttore chiamato per myv[2]  ← ordine inverso!
  → distruttore chiamato per myv[1]
  → distruttore chiamato per myv[0]
```

````python
{
    std::cout << "=== Inizio: array di 3 Foo ==="  << std::endl;
    Foo myv[3];  // costruttore chiamato 3 volte
    std::cout << "=== Fine scope: distruttori in ordine INVERSO ==="  << std::endl;
    // uscita dallo scope → 3 distruttori
}
````

## 9. `std::vector` di Oggetti

`std::vector<Foo> v(12)` crea un vector con **12 oggetti** `Foo`, ognuno inizializzato con il costruttore di default.

Il vantaggio rispetto all'array C-style:
- Dimensione **dinamica** (può crescere con `push_back`)
- Gestione della memoria **automatica** (nessun `delete` manuale)
- Quando il vector viene distrutto, chiama il distruttore su **tutti** gli elementi

````python
{
    std::cout << "=== Vector di 3 Foo ==="  << std::endl;
    std::vector<Foo> v(3);  // costruttore di default chiamato 3 volte
    v[0].setX(10);
    v[1].setX(20);
    v[2].setX(30);
    for (int i = 0; i < 3; i++) v[i].print();
    std::cout << "=== Fine scope: vector distrutto ==="  << std::endl;
    // vector esce dallo scope → 3 distruttori automatici
}
````

## 10. ⚠️ Passaggio per Valore → Segmentation Fault!

Questa è la parte più importante. Considera questa funzione:

```cpp
void funzione(Foo a) {  // 'a' è una COPIA di x
    a.print();
}  // ← distruttore di 'a' chiamato qui: delete x

Foo x;
funzione(x);  // passa una copia di x
// x.~Foo() chiamato alla fine → tenta delete x... ma è già stata deallocata! 💥
```

### Perché crasha?

Il compilatore, non avendo un **copy constructor** definito, fa una **shallow copy**:
copia solo il valore del puntatore `x`, non la memoria a cui punta.

```
SHALLOW COPY (sbagliata):

 x (originale)           a (copia locale)
 ┌─────────────┐         ┌─────────────┐
 │  x → 0xABCD │         │  x → 0xABCD │  ← stesso indirizzo!
 └─────────────┘         └─────────────┘
                                │
                  fine funzione → delete 0xABCD  ← memoria liberata!

 poi x esce dallo scope → delete 0xABCD  ← DOUBLE FREE → SEGFAULT 💥
```

### La soluzione: il Copy Constructor
Con il **copy constructor** (vedi notebook precedente), la copia alloca **nuova memoria**:

```
DEEP COPY (corretta):

 x (originale)           a (copia locale)
 ┌─────────────┐         ┌─────────────┐
 │  x → 0xABCD │         │  x → 0xEF01 │  ← memoria DIVERSA!
 └─────────────┘         └─────────────┘
        │                        │
      [42]                     [42]     ← stesso VALORE, indirizzi separati
```

> 🔑 Questo è il motivo della **Regola del Tre**: se hai un distruttore che fa `delete`,
> hai **bisogno** anche del copy constructor (e dell'operatore di assegnazione).

````python
// ATTENZIONE: questa cella dimostra il problema.
// Senza copy constructor, il passaggio per valore causa double-free.
// La cella è lasciata commentata per sicurezza — decommentare solo per vedere il crash.

/*
void funzioneUnsafe(Foo a) {  // shallow copy: a.x punta alla stessa memoria di x.x
    a.print();
}  // ← delete a.x: libera la memoria condivisa!

Foo x;
x.setX(42);
funzioneUnsafe(x);  // a.~Foo() libera la memoria
x.print();          // x.x è un dangling pointer → UNDEFINED BEHAVIOR
// x.~Foo() → double free → SEGMENTATION FAULT 💥
*/
````

###  Soluzione  Passaggio per Riferimento `const&`

Se la funzione non deve modificare l'oggetto, usare `const Foo&` è la scelta ottimale:
- **Nessuna copia** → nessun copy constructor → nessun distruttore intermedio
- **Più efficiente** per oggetti grandi
- `const` garantisce che la funzione non modifichi l'originale

````python
void funzioneRef(const FooSafe& a) {  // alias: nessuna copia!
    a.print();
}  // nessun distruttore — 'a' non è un oggetto reale, è solo un alias

{
    FooSafe obj;
    obj.setX(99);
    std::cout << "Passaggio per riferimento: ";
    funzioneRef(obj);  // ✅ nessuna copia, nessun overhead
}
````

##  Riepilogo Finale

| Situazione | Costruttore chiamato | Distruttore chiamato |
|---|---|---|
| `Foo f;` | Default | Fine scope |
| `Foo arr[N];` | Default × N | Fine scope × N (inverso) |
| `Foo* p = new Foo;` | Default | Solo con `delete p` |
| `std::vector<Foo> v(N)` | Default × N | Distruzione del vector |
| `void f(Foo a)` chiamata | **Copy constructor** | Fine funzione |
| `void f(Foo& a)` chiamata | Nessuno | Nessuno |

###  La Trappola classica
Se hai `delete` nel distruttore ma **non hai il copy constructor** → la shallow copy del compilatore
porta a **double free** → **Segmentation Fault**.

>  **Regola del Tre**: se definisci uno tra `~Foo()`, `Foo(const Foo&)`, `operator=`, definiscili tutti e tre.
