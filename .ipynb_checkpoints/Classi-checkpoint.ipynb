```c++
class Foo{
    public:
        Foo();
        ~Foo();
        void print();
        void setX(int a);
    private:
        int * x;
};
```

La classe foo è costituita da una parte publlica è una privata. Nella parte publica ci sono due metodi: 
- **print**: stampa lo stato della classe;
- **set**: consente di impostare lo stato della classe.
all'interno della classe abbiamo fatto solo la dichiarazione dei metodi mentre all'esterno abbiamo fatto l'implementazione dei metodi.

Costruttore di default: 

```c++
 Foo::Foo(){
     x=new int;
    *x= 0; 
    std::cout<<"Chiamato costruttore di default"<<std::endl;
}
```

Distruttore:

```c++
Foo::~Foo(){
    std::cout<<"Distruttore di classe con x= "<<* x<<std::endl;
    delete x;
}
```

```c++
void Foo::print(){
    std::cout<<x<<std::endl;
}
```

```c++
void Foo::setX(int a){
    *x= a; 
}
```
